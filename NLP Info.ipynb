#### Text Mining or Text Analytics - Process of deriving meaningful information from natural language text - Involves process of structuring the input text, deriving patterns within the structured data and finally evaluating & interpreting the output.

#### Natural Language Processing refers to the AI method of communicating with an intelligent system using natural language

#### The main goal is to turn the text into data analysis via the application of NLP, so in turn NLP & Text mining/analytics go hand in hand

##### Applications of NLP:
    - Sentiment Analysis
    - Chatbots 
    - Speech recognition in systems like Siri, Cortana & Alexa
    - Machine Translation - Google or MS Translate
    - Spell check
    - Keyword search
    - Information extraction
    - Advertisement matching - Ad recomendations based on your search history

#### Components of NLP
1. Natural language Understanding - Mapping the given input into natural language into useful representation and analysing those aspects of the language 
2. Natural language Generation - Process of producing meaningful phrases & sentences in the form of natural language from some internal representation

#### Various steps involved in NLP :
1. Tokenisation - Breaking strings into tokens  
2. Stemming - Normalising the words into its base or root form : Eg. Affected or Affections converted to Affect so removing the beginning or end of the word 
3. Lemmatization - Takes into consideration the morphological analysis of the word so as to link the form back to its original word, groups together different inflected forms of a word or lemma difference between Stemming & Lemmitization is the O/P of lemmitization is always a word : Eg. gone, going & went are linked back to "go"
4. POS(Part of Speech) Tags - After the tokens have been converted into its root forms, then POS comes into the picture, grammatical type of the word is referred to as the POS tag, be it the verb, noun, adjective or article

5. Named Entity Recognition - Detecting named entities such as person or company names, quantities or the locations, 3 Steps - Noun phrase identification, phrase classification & entity disambiguation:
